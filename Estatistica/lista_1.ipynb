Problema 5: Um artigo em Journal of Aircraft descreve o calculo computacional de coeficientes de arrasto
(CD) para o aerofoloio NACA0012. Diferentes algoritmos computacionais foram utilizados em
M∞ = 0,7, com os seguintes resultados (para obter os valores verdadeiros do CD, multiplicar
os dados por 10^4): 79, 100, 74, 83, 81, 85, 82, 80, 84.

(a) Ordene os dados em ordem crescente;

(b) Faca um diagrama de pontos desses dados;

(c) Calcule a media e a mediana. Inclua essas informacoes no diagrama de pontos;

(d) Substitua o valor 100 por 150. O que acontece com media e mediana?

(e) Calcule a variancia e o desvio-padrao (com os dados originais);

(f) Subtraia 10 de cada uma das medidas originais e calcule novamente a variancia. Qual
conclusao se pode tirar?

(g) Multiplique os dados originais por 10 e calcule novamente a variancia e o desvio-padrao.
Quais conclusoes se podem tirar?

```python
import matplotlib.pyplot as plt
import numpy as np
vet1=[79, 100, 74, 83, 81, 85, 82, 80, 84]

#A:
print("\nA:\n")
vet1.sort()
print(vet1)

#B:
print("\nB:\n")
plt.figure(figsize=(8, 3))
plt.plot(vet1, np.zeros_like(vet1), 'o', label="Dados")
plt.title('Diagrama de Pontos dos Coeficientes de Arrasto (CD)')
plt.yticks([])  # Remover marcação do eixo Y
plt.show()

#C:
print("\nC:\n")
print("Media:", bf.media(vet1))
print("Mediana:",bf.mediana(vet1))

plt.figure(figsize=(8, 3))
plt.plot(vet1, np.zeros_like(vet1), 'o', label="Dados")

# Adicionar linha da média e da mediana
plt.axvline(bf.media(vet1), color='r', linestyle='--', label=f'Média: {bf.media(vet1):.2f}')
plt.axvline(bf.mediana(vet1), color='g', linestyle='--', label=f'Mediana: {bf.mediana(vet1):.2f}')

# Customizar o gráfico
plt.title('Diagrama de Pontos dos Coeficientes de Arrasto (CD)')
plt.yticks([])  # Remover marcação do eixo Y
plt.legend()    # Exibir a legenda
plt.show()

#D:
print("\nD:\n")
for i in range(len(vet1)): 
    if(vet1[i]==100):vet1[i]=150
print("\nVetor alterado:", vet1)
print("Media:", bf.media(vet1))
print("Mediana:",bf.mediana(vet1))

#E:
print("\nE:\n")
for i in range(len(vet1)): 
    if(vet1[i]==150):vet1[i]=100
print("\nVetor alterado:", vet1)
print("Variancia:", bf.variancia(vet1))
print("Desvio padrao:",bf.desvio_padrao(vet1))

#F:
print("\nF:\n")
for i in range(len(vet1)): vet1[i]-=10
print("\nVetor alterado:", vet1)  
print("Variancia:", bf.variancia(vet1))

#G:
print("\nG:\n")
for i in range(len(vet1)): vet1[i]=(vet1[i]+10)*10
print("\nVetor alterado:", vet1)  
print("Variancia:", bf.variancia(vet1))
print("Desvio padrao:",bf.desvio_padrao(vet1))
```
